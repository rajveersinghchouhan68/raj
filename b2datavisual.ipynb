```
x=[2,5,4,9]
y=[3,8,3,6]
import matplotlib.pyplot as plt
plt.xlabel('distance')   #for labelling x axis
plt.ylabel('time')       #for y
plt.scatter(x,y,label="mins",color='green')         # to show dots
plt.scatter(y,x)
plt.plot(x,y,label="railway")            # to plot line
plt.plot(x,y,color='lightgreen',label="financegrowth")            # to show bar plot
plt.bar(y,x,color='red')
plt.scatter([2,5],[3,8])
plt.plot(x,y)            #to show bar plot 
plt.grid(c='red')        #for grid and color as well
plt.plot()               #to plot the graph
```
